```
import math

# Define the variables and titles for plotting
plots_info = [
    ('GPA', 'GPA by Admission Decision'),
    ('TestScore', 'Test Score by Admission Decision'),
    ('WorkExp', 'Work Experience by Admission Decision'),
    ('WritingScore', 'Writing Score by Admission Decision'),
    ('VolunteerLevel', 'Volunteer Level by Admission Decision')
]

num_plots = len(plots_info)
ncols = 2
nrows = math.ceil(num_plots / ncols)

# Create subplots with 2 plots per row
fig, axs = plt.subplots(nrows, ncols, figsize=(15, nrows * 5))
axs = axs.flatten()  # Flatten the array to iterate easily

# Create each boxplot
for i, (col, title) in enumerate(plots_info):
    sns.boxplot(x='Decision', y=col, data=data, ax=axs[i])
    axs[i].set_title(title)

# Remove any extra subplots (if total plots is odd, one subplot will be empty)
for j in range(num_plots, len(axs)):
    fig.delaxes(axs[j])

plt.tight_layout()
plt.savefig('combined_boxplots_2_per_row.png')
plt.show()
```
